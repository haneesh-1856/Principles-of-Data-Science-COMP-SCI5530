```python
import pandas as pds
import matplotlib.pyplot as plt
import seaborn as sns
```

1) Look for the missing values in all the columns and either impute them (replace with mean, median, or mode) or drop them. Justify your action for this task.

```python
def extract_numeric(data_Set):
    return pds.to_numeric(data_Set.astype(str).str.extract(r'([\d.]+)')[0], errors='coerce')
dset_num=pds.DataFrame(columns=[])
dset_num["Mileage"]=extract_numeric(dset["Mileage"])
dset_num["Engine"]=extract_numeric(dset["Engine"])
dset_num["Power"]=extract_numeric(dset["Power"])
print(dset_num)
dset["Mileage"].fillna(dset_num["Mileage"].mean(), inplace=True)
dset["Engine"].fillna(dset_num["Engine"].mean(), inplace=True)
dset["Power"].fillna(dset_num["Power"].mean(), inplace=True)
dset["Seats"].fillna(dset["Seats"].mode()[0], inplace=True)
dset["Fuel_Type"].fillna(dset["Fuel_Type"].mode()[0], inplace=True)
dset["Transmission"].fillna(dset["Transmission"].mode()[0], inplace=True)
dset["Owner_Type"].fillna(dset["Owner_Type"].mode()[0], inplace=True)

dset.drop(columns=["New_Price"], inplace=True)
dset
```

**Justity**: All the Numerice Data Type colummn null values have been replaced with its mean. Categorical values are replaced with their columns mode value. New_Price is a numeric variable and the data set have only 815 entries for the "New_Price" column out of 5847, which is mostly null. Also prices vary greatly and often aren't avaliable for older models. So, the "New_Price" column is dropped off the dataset.

2) Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”).

```python
dset["Mileage"] = dset["Mileage"].astype(str).str.extract(r'([\d\.]+)').astype(float)
dset["Engine"] = dset["Engine"].astype(str).str.extract(r'([\d\.]+)').astype(float)
dset["Power"] = dset["Power"].astype(str).str.extract(r'([\d\.]+)').astype(float)
#dset["New_Price"] = dset["New_Price"].astype(str).str.extract(r'([\d\.]+)').astype(float)
dset
```

3) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value.

If the Fuel_type is "Petrol" then the values will be 1, and 2 for "Diesel" Fuel_type

If the Transmission type is "Manual" it's give value 1, and 2 for the Transmission type "Automatic"

```python
dset["Fuel_Type"] = dset["Fuel_Type"].apply(lambda x: 1 if x.lower()=="petrol" else 2)
dset["Transmission"] = dset["Transmission"].apply(lambda x: 1 if x.lower()=="manual" else 2)
dset
```
